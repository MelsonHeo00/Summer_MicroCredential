## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
"""
Description: A program where the graph is matched based on the execrise 4 criteria
Created: Melson Heo
Input: None
Output: A graph matching the graph below
"""

flg, axes = plt.subplots(nrows = 1, ncols = 2) # Creates 2 sub plots that features one row and 2 columns
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
fig.tight_layout()
ax1.plot(x,y,'b--', linewidth = 3)
ax2.plot(x,z,'r', linewidth = 3)

ax1.set_xlim([0,100]) # Limits the graph for the x axis
ax2.set_xlim([0,100]) # Limits the graph for the x axis
ax1.set_ylim([0,200]) # Limits the graph for the y axis
ax2.set_ylim([0,10000]) # Limits the graph for the y axis
```

*See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.*

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (10.1,2))
fig.tight_layout()
ax1.plot(x,y,'b', linewidth = 5)
ax2.plot(x,z,'r--', linewidth = 2)

ax1.set_xlim([0,100]) # Limits the graph for the x axis
ax2.set_xlim([0,100]) # Limits the graph for the x axis
ax1.set_ylim([0,200]) # Limits the graph for the y axis
ax2.set_ylim([0,10000]) # Limits the graph for the y axis
```
